```python
import csv
 
f = open('KRvideos.csv', 'r', encoding='utf-8')
data = csv.reader(f)

dic = {}
    
for i in range(34600):
    try:
        line = next(data)
    except:
        continue
    
    if i == 0:
        columns = line.copy()
        for col in columns:
            dic[col] = []
        continue
        
    if len(line) != 16:
        continue
        
    for j, elem in enumerate(line):

        dic[columns[j]].append(elem)
        
f.close()
```

```python
df = pd.DataFrame(dic)
df.tail()
```

```python
type_dict = {}
type_dict[columns[0]] = str
type_dict[columns[1]] = str
type_dict[columns[2]] = str
type_dict[columns[3]] = str
type_dict[columns[4]] = int
type_dict[columns[5]] = str
type_dict[columns[6]] = str
type_dict[columns[7]] = int
type_dict[columns[8]] = int
type_dict[columns[9]] = int
type_dict[columns[10]] = int
type_dict[columns[11]] = str
type_dict[columns[12]] = bool
type_dict[columns[13]] = bool
type_dict[columns[14]] = bool
type_dict[columns[15]] = str
```

```python
type_dict
```

```python
for col in columns:
    df[col] = df[col].apply(type_dict[col])
```

```python
with open("KR_category_id.json", "r") as f:
    cate_info = json.loads(f.read())
```

```python
len(df.category_id.unique())
```

```python
category_id_name = {}

for cate in cate_info["items"]:
    
    category_id_name[int(cate["id"])] = cate["snippet"]["title"]
    
    
category_id_name[29] = "Nonprofits & Activism" ## it is missed, this is from US version json file.
category_id_name
```

```python
df["category_name"] = list(map(lambda x : category_id_name[x], df.category_id.values))
```
