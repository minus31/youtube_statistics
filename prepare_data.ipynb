## Youtube statistics 

 - data source : https://www.kaggle.com/datasnaek/youtube-new/data#KRvideos.csv
 
 --- 
  위 링크에서 제공하는 데이터 중, 한국에 해당하는 것은 중간중간 망가진 데이터가 있어서 일반적인 방법으로 읽을 수 없었다. 따라서 다음과 같은 과정을 통해 데이터를 정제 하였다. 
  
  1. Python의 csv 라이브러리를 사용하여 데이터의 generator를 생성한다. 
  2. next()를 호출했을 때, 에러가 없다면 다음 과정을 수행한다. 
  3. 가장 첫번째 데이터 행은 해당 데이터의 column을 의미한다. 
  4. 데이터의 요소 갯수가 맞지 않는다면, 그 데이터는 제외하도록 한다. 
  5. 데이터를 string 타입으로 생성하고 이를 각 column의 리스트에 append한다. 
  6. 각 column의 데이터타입을 손수 정의한다. 

```python
import csv
 
f = open('KRvideos.csv', 'r', encoding='utf-8')
data = csv.reader(f)

dic = {}
    
for i in range(34600):
    try:
        line = next(data)
    except:
        continue
    
    if i == 0:
        columns = line.copy()
        for col in columns:
            dic[col] = []
        continue
        
    if len(line) != 16:
        continue
        
    for j, elem in enumerate(line):

        dic[columns[j]].append(elem)
        
f.close()
```

29의 카테고리에 대해서는 누락이 되어 있다. 이는 미국의 자료를 참고하여 손수 정하였다. 전체를 훓어봤을 때, 한국과 미국의 카테고리id와 이름은 똑같았다. 

```python
category_id_name = {}

for cate in cate_info["items"]:
    
    category_id_name[int(cate["id"])] = cate["snippet"]["title"]
    
    
category_id_name[29] = "Nonprofits & Activism" ## it is missed, this is from US version json file.
category_id_name
```

```python
df["category_name"] = list(map(lambda x : category_id_name[x], df.category_id.values))
```
